**Q1. What is the difference between a function and a method in Python?**

The main difference between a **function** and a **method** in Python lies in how they are used and where they belong:

1. **Function**  
   - A function is a **standalone block of reusable code** that performs a specific task.  
   - It is defined using the `def` keyword.  
   - It can take arguments and return values.  
   - It can be called independently, without being associated with any object.  

   **Example of a function:**  
   ```python
   def greet(name):
       return f"Hello, {name}!"

   print(greet("Deepanshu"))  # Output: Hello, Deepanshu!
   ```

2. **Method**  
   - A method is a **function that is associated with an object** (usually part of a class).  
   - It operates on an instance of the class and typically acts on its attributes.  
   - It is called using the **dot notation** (`object.method()`).  
   - The first parameter of an instance method is always `self`, referring to the object itself.  

   **Example of a method inside a class:**  
   ```python
   class Person:
       def __init__(self, name):
           self.name = name
       
       def greet(self):
           return f"Hello, {self.name}!"

   person = Person("Deepanshu")
   print(person.greet())  # Output: Hello, Deepanshu!
   ```

**Q2. Explain the concept of function arguments and parameters in Python.**


In Python, **parameters** and **arguments** are terms related to functions, but they have distinct meanings.

---

**1. Parameters**  
- **Definition:** Variables **defined in the function signature** (inside parentheses) that receive values when the function is called.  
- **Purpose:** They act as placeholders for values that the function will process.  

**Example:**  
```python
def greet(name):  # 'name' is a parameter
    return f"Hello, {name}!"
```
---

**2. Arguments**  
- **Definition:** Actual **values passed to the function** when calling it.  
- **Purpose:** They provide real data for the function to work with.  

**Example:**  
```python
print(greet("Deepanshu"))  # "Deepanshu" is an argument
```

---

**Types of Function Arguments in Python**  

Python supports multiple types of function arguments:

**1. Positional Arguments**  
- Values are assigned to parameters based on their position in the function call.  
- The order matters.  

**Example:**  
```python
def add(a, b):
    return a + b

print(add(5, 10))  # Output: 15
```

---

**2. Default Arguments**  
- If an argument is **not provided**, the default value is used.  
- Helps make functions more flexible.  

**Example:**  
```python
def greet(name="Guest"):
    return f"Hello, {name}!"

print(greet())         # Output: Hello, Guest!
print(greet("Alex"))   # Output: Hello, Alex!
```

---

**3. Keyword Arguments (Named Arguments)**  
- Arguments are passed using **parameter names**, not just positions.  
- Order does **not** matter.  

**Example:**  
```python
def introduce(name, age):
    return f"My name is {name} and I am {age} years old."

print(introduce(age=25, name="Deepanshu"))  
# Output: My name is Deepanshu and I am 25 years old.
```

---

**4. Variable-Length Arguments (`*args` and `**kwargs`)**  
Python allows passing an arbitrary number of arguments using:

**(i) `*args` (Non-keyword Arguments)**
- Allows passing multiple values as a **tuple**.  

**Example:**  
```python
def add_numbers(*args):
    return sum(args)

print(add_numbers(2, 4, 6, 8))  # Output: 20
```

---

**(ii) `**kwargs` (Keyword Arguments)**
- Allows passing multiple **named** arguments as a **dictionary**.  

**Example:**  
```python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Deepanshu", age=25, city="Delhi")
```
**Output:**
```
name: Deepanshu
age: 21
city: Delhi
```

 **Q.5 What are iterators in Python and how do they differ from iterables?**


**Iterators vs. Iterables in Python**  

**Iterable:**  
- An object that contains multiple elements and can be looped over.  
- Has the `__iter__()` method, which returns an iterator.  
- Examples: Lists, tuples, dictionaries, sets, strings.  

```python
numbers = [1, 2, 3]  # List is an iterable
for num in numbers:
    print(num)
```

**Iterator:**  
- An object that produces elements one at a time.  
- Has `__iter__()` and `__next__()` methods.  
- Stores state and returns the next value using `next()`.  
- Raises `StopIteration` when no more elements are left.  

```python
numbers = iter([1, 2, 3])  # Convert list to iterator
print(next(numbers))  # Output: 1
print(next(numbers))  # Output: 2
print(next(numbers))  # Output: 3
print(next(numbers))  # Raises StopIteration
```

**Key Differences:**  
| Feature      | Iterable | Iterator |
|-------------|---------|---------|
| Methods     | `__iter__()` | `__iter__()`, `__next__()` |
| Stores Elements? | Yes | No, remembers the current state |
| Usage      | Used in loops (`for`, `while`) | Used with `next()` |

**Custom Iterator Example:**  
```python
class Counter:
    def __init__(self, limit):
        self.count = 0
        self.limit = limit

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.limit:
            self.count += 1
            return self.count
        raise StopIteration

counter = Counter(3)
for num in counter:
    print(num)  # Output: 1, 2, 3
```


**Q6. Explain the concept of generators in Python and how they are defined.**


**What is a Generator?**  
- A **generator** is a special type of iterator that **yields** values one at a time using the `yield` keyword instead of `return`.  
- It does not store all values in memory but generates them on the fly, making it **memory-efficient**.  
- Generators are used for **large datasets** and **streaming data** where storing all elements at once is impractical.  

---

**Defining a Generator**  
A generator is defined using a function with `yield`.  

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()  # Creates a generator object
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
print(next(gen))  # Raises StopIteration
```

---

**How Generators Work?**  
- When `next()` is called, execution resumes from where it left off.  
- The function pauses at `yield`, **saves its state**, and resumes on the next call.  
- Once all values are yielded, it raises `StopIteration`.  

---

**Generator vs. Normal Function**  

| Feature        | Generator (`yield`) | Normal Function (`return`) |
|---------------|---------------------|----------------------------|
| Memory Usage  | Efficient, stores only current state | Stores all values at once |
| Execution     | Pauses and resumes | Runs and exits completely |
| Output Type   | Iterator (can use `next()`) | Direct return value |

---

**Generator Example: Infinite Sequence**  
```python
def count_up(start):
    while True:
        yield start
        start += 1

counter = count_up(1)
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
```
Useful for streaming or continuous data generation.

---

**Using Generators in a Loop**  
Generators can be used directly in loops without calling `next()`.  

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(3):
    print(num)  # Output: 3, 2, 1
```

---

**Generator Expression (Like List Comprehension but Memory Efficient)**  
```python
gen = (x**2 for x in range(5))
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
```
Consumes less memory than a list comprehension `squares = [x**2 for x in range(5)]`.

---

**Conclusion**  
- **Generators use `yield` instead of `return`** to produce values lazily.  
- **They save memory** by generating values on demand.  
- **Can be infinite**, making them useful for streams and large datasets.  
- **Generator expressions** provide a compact way to create generators.

**Q7. What are the advantages of using generators over regular functions?**


**1. Memory Efficiency**  
- Generators **do not store** all values in memory; they generate values **on demand**.  
- Useful for handling **large datasets** and **streams of data**.  

```python
def large_numbers():
    for i in range(10**6):
        yield i  # Generates one value at a time

gen = large_numbers()  # Does not consume large memory
```

---

**2. Lazy Evaluation (On-Demand Execution)**  
- Generators pause execution at `yield`, resuming when `next()` is called.  
- This **avoids unnecessary computations** and speeds up execution.  

```python
def lazy_range(n):
    for i in range(n):
        yield i  # Generates values when needed

gen = lazy_range(5)
print(next(gen))  # Output: 0 (computation continues only when needed)
```

---

**3. Faster Execution (No Overhead of Storing Data)**  
- Regular functions store results in lists, consuming more **time and space**.  
- Generators **avoid overhead** by yielding results **as needed**.  

```python
def square_list(n):
    return [i**2 for i in range(n)]  # Stores all values in memory

def square_generator(n):
    for i in range(n):
        yield i**2  # Generates one value at a time
```

---

**4. Infinite Sequences**  
- Generators can produce **infinite values** without exhausting memory.  
- Useful for **stream processing**, **sensor data**, etc.  

```python
def count_up(start):
    while True:
        yield start
        start += 1  # Keeps generating numbers infinitely

counter = count_up(1)
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
```

---

**5. Simplified Code (No Need to Manage State Manually)**  
- With normal functions, we need to manage **loops, lists, and state tracking**.  
- Generators **simplify logic** with `yield`, making code **cleaner**.  

```python
def even_numbers(n):
    for i in range(0, n, 2):
        yield i  # No need for extra storage
```

---

**6. Generator Expressions (Compact & Efficient)**  
- Generators can be defined in **one line** using **generator expressions**.  

```python
gen = (x**2 for x in range(5))
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
```
- This is more **memory-efficient** than list comprehensions:  
  ```python
  lst = [x**2 for x in range(5)]  # Stores all values in memory
  ```

**8. What is a lambda function in Python and when is it typically used?**

**Lambda Function in Python**  

A **lambda function** in Python is an **anonymous, single-expression function** defined using the `lambda` keyword. It is typically used for **short, simple operations** where defining a full function is unnecessary.  

---

**Syntax of Lambda Function**  
```python
lambda arguments: expression
```
- **No `def` keyword or function name**  
- **Can have multiple arguments but only one expression**  
- **Returns the result automatically**  

---

**Example: Basic Lambda Function**  
```python
square = lambda x: x ** 2
print(square(5))  # Output: 25
```

Equivalent to:
```python
def square(x):
    return x ** 2
```

---

**When to Use Lambda Functions?**  

**1. Inside `map()`, `filter()`, and `reduce()`**  
Used for quick transformations on data.  

```python
numbers = [1, 2, 3, 4]

# Double each number using map()
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)  # Output: [2, 4, 6, 8]

# Filter even numbers using filter()
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4]
```

---

**2. Sorting with Custom Keys**  
Used as a key function in sorting.  

```python
students = [("Alice", 25), ("Bob", 22), ("Charlie", 23)]
students.sort(key=lambda x: x[1])  # Sort by age
print(students)  # Output: [('Bob', 22), ('Charlie', 23), ('Alice', 25)]
```

---

**3. Inside Higher-Order Functions**  
When passing simple functions as arguments.  

```python
def apply_func(f, value):
    return f(value)

print(apply_func(lambda x: x + 10, 5))  # Output: 15
```

---

**Conclusion**  
- **Lambda functions** are best for **short, throwaway functions**.  
- Useful in **functional programming** (`map()`, `filter()`, `sorted()`).  
- Should be avoided for **complex logic** for better readability.

**Q9.  Explain the purpose and usage of the 'map()' function in Python?
  

**Purpose**  
The `map()` function is used to **apply a function to each item** in an iterable (like a list or tuple) and return a new iterable with transformed values. It is useful for performing element-wise operations **without using explicit loops**.

---

**Syntax**  
```python
map(function, iterable)
```
- **`function`** → A function that is applied to each element.  
- **`iterable`** → A sequence (list, tuple, etc.) to be processed.  
- Returns a **map object (iterator)** that can be converted to a list or tuple.  

---

**Example: Using `map()` with a Regular Function**  
```python
def square(n):
    return n ** 2

numbers = [1, 2, 3, 4]
result = map(square, numbers)  # Applies square() to each element
print(list(result))  # Output: [1, 4, 9, 16]
```

---

**Example: Using `map()` with a Lambda Function**  
```python
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
```
**Lambda makes `map()` more concise**.

---

**Example: Using `map()` with Multiple Iterables**  
You can pass **multiple iterables** to `map()`, and the function should accept that many arguments.

```python
a = [1, 2, 3]
b = [4, 5, 6]

sum_values = list(map(lambda x, y: x + y, a, b))
print(sum_values)  # Output: [5, 7, 9]
```

---

**Using `map()` with `str` Methods**  
```python
words = ["hello", "world", "python"]
uppercase = list(map(str.upper, words))
print(uppercase)  # Output: ['HELLO', 'WORLD', 'PYTHON']
```
**Directly applies `str.upper` method to all elements**.

---

**When to Avoid `map()`?**  
When **list comprehensions** provide a more readable alternative:  

```python
numbers = [1, 2, 3, 4]
squared = [x ** 2 for x in numbers]  # More Pythonic

```
 When functions require **complex logic**, as `map()` is best for simple transformations.  

---

### **Conclusion**  
- `map()` applies a function to every item in an iterable efficiently.  
- Works well with **lambda functions** for quick transformations.  
- Use when **transforming elements**, but prefer **list comprehensions** for readability.

**Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

These functions are part of **functional programming** in Python and operate on iterables like lists and tuples.

---

**1. `map()` – Applies a Function to Each Element**  
- **Purpose:** Transforms each element of an iterable using a given function.  
- **Returns:** A `map` object (iterator) containing transformed values.  

**Example: Square each number**  
```python
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
```

**Use `map()` when you need to modify all elements.**  

---

**2. `filter()` – Filters Elements Based on a Condition**  
- **Purpose:** Selects elements that satisfy a condition (returns `True`).  
- **Returns:** A `filter` object (iterator) containing filtered elements.  

**Example: Keep only even numbers**  
```python
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
```

**Use `filter()` when you need to remove elements that don't match a condition.**  

---

### **3. `reduce()` – Reduces an Iterable to a Single Value**  
- **Purpose:** Repeatedly applies a function to **accumulate** a result.  
- **Returns:** A **single computed value**.  
- **Requires:** `functools.reduce()` module.  

**Example: Find the product of all numbers**  
```python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24
```

**Use `reduce()` when you need to aggregate values into a single result (sum, product, max, etc.).**  

---

**Comparison Table**  

| Function  | Purpose | Returns | Best Use Case |
|-----------|---------|---------|---------------|
| `map()`   | Applies a function to each element | Transformed iterable | When modifying all elements |
| `filter()` | Filters elements based on a condition | Filtered iterable | When selecting certain elements |
| `reduce()` | Reduces elements to a single value | Single value | When aggregating (sum, product, etc.) |

---

**Conclusion**  
- **Use `map()`** when transforming **all elements**.  
- **Use `filter()`** when selecting **specific elements**.  
- **Use `reduce()`** when aggregating **to a single value**.  
- **For better readability**, use **list comprehensions** instead of `map()` and `filter()` when possible.

````python
#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200

def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to √n
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
primes = [num for num in range(1, 201) if is_prime(num)]
print(primes)
````

````python
#Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Track iterations

    def __iter__(self):
        return self  # Returns itself as an iterator

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # Stop when limit is reached
        fib = self.a
        self.a, self.b = self.b, self.a + self.b  # Update Fibonacci numbers
        self.count += 1
        return fib

fib_iter = FibonacciIterator(10)
print(list(fib_iter))
````

````python
#Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):  # Loop from 0 to exponent
        yield 2 ** i  # Yield power of 2

for power in powers_of_2(5):
    print(power)
````

````python
#Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    vowels = "aeiouAEIOU"  # Define vowels
    return "".join(filter(lambda ch: ch not in vowels, s))  # Filter out vowels

# Example usage
text = "Hello, World!"
result = remove_vowels(text)
print(result)  # Output: "Hll, Wrld!"
````

````python
# Q.11 Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and theproduct of the price per item and the quantity.  The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

def calculate_order_total(orders):
    result = []
    for order in orders:
        order_number, price_per_item, quantity = order
        total_cost = price_per_item * quantity
        if total_cost < 100:
            total_cost += 10  # Add 10€ if the total is less than 100€
        result.append((order_number, total_cost))
    return result

# Example usage
orders = [
    (1, 20, 3),  # 20 * 3 = 60 < 100, so 60 + 10 = 70
    (2, 15, 7),  # 15 * 7 = 105, no extra charge
    (3, 50, 1)   # 50 * 1 = 50 < 100, so 50 + 10 = 60
]

print(calculate_order_total(orders))


````
